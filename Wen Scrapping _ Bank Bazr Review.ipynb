```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.bankbazaar.com/reviews.html'
```

```python
page = requests.get(url)
```

```python
page.text
```

```python
#parse the above html
#beautiful soup is used for parsing html as well xml parser both
soup = BeautifulSoup(page.text,'html.parser')
```

```python
soup
```

```python
review_element = soup.find_all(class_ = 'text_here review-desc-more')
```

```python
review_element
```

```python
sentiment = []
for i in review_element:
    sentiment.append(i.text)
```

```python
sentiment
```

```python
sentiment = list(map(lambda x:x.strip(),sentiment))
```

```python
sentiment
```

```python
ratings = soup.find_all(class_='rating-section review-user-score')
ratings
```

```python
score=[]
for item in ratings:
    score.append(item.text)
score
```

```python
response=[]
for i in score:
    response.append(i.split()[1])
```

```python
response
```

```python
Rating=[]
for i in response:
    Rating.append(int(i[0]))
```

```python
Rating
```

```python
feedback=[]
for i in score:
    feedback.append(i.split()[2])
```

```python
feedback
```

```python
comment = soup.find_all(class_='user-review-comment js-individual-title')
comment
```

```python
summary = []
for i in comment:
    summary.append(i.text)
```

```python
summary
```

```python
name  = soup.find_all(class_='review-bank-title')
```

```python
bank = []
for i in name:
    bank.append(i.find('img').get('title'))
```

```python
bank
```

```python
final = []
for i,j,k,l,m in zip(bank,feedback,sentiment,summary,Rating):
    final.append({'Bank':i,'Feedback':j,'Sentiment':k,'Summary':l,'Rating':m})
```

```python
df = pd.DataFrame(final)
df.head()
```

```python
df.info()
```

```python
import nltk
nltk.download()
```

```python
df = df.drop(['Bank','Feedback','Summary'],axis=1)
```

```python
df.head()
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
lemma = WordNetLemmatizer()
```

```python
corpus = []

for i in range(0,len(df)):
    words = re.sub('[^a-zA-Z]',' ',df['Sentiment'][i])
    words = words.lower()
    words = words.split()
    words = [lemma.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    words = ' '.join(words)
    corpus.append(words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tv = TfidfVectorizer()
```

```python
x = tv.fit_transform(corpus)
```

```python
y = df['Rating']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
import pickle
import sklearn
import scipy
```

```python
def naive_classifier(x_train, x_test, y_train, y_test):
    
    classifier_naive = MultinomialNB()
    classifier_naive.fit(x_train, y_train)

    y_pred = classifier_naive.predict(x_test)

    
    cm = confusion_matrix(y_test, y_pred)

    return print(f"Train score : {classifier_naive.score(x_train, y_train)}\nTest score : {classifier_naive.score(x_test, y_test)}")
```

```python
def random_classifier(x_train, x_test, y_train, y_test):
    
    classifier_rf = RandomForestClassifier()
    classifier_rf.fit(x_train, y_train)

    y_pred = classifier_rf.predict(x_test)

    cm = confusion_matrix(y_test, y_pred)

    return print(f"Train score : {classifier_rf.score(x_train, y_train)}\nTest score : {classifier_rf.score(x_test, y_test)}")
```

```python
def ada_classifier(x_train, x_test, y_train, y_test):
    
    classifier_ab = AdaBoostClassifier()
    classifier_ab.fit(x_train, y_train)

    y_pred = classifier_ab.predict(x_test)

    cm = confusion_matrix(y_test, y_pred)

    return print(f"Train score : {classifier_ab.score(x_train, y_train)}\nTest score : {classifier_ab.score(x_test, y_test)}")
```

```python
def knn_classifier(x_train, x_test, y_train, y_test):
    
    classifier_knn = KNeighborsClassifier()
    classifier_knn.fit(x_train, y_train)

    y_pred = classifier_knn.predict(x_test)

    cm = confusion_matrix(y_test, y_pred)

    return print(f"Train score : {classifier_knn.score(x_train, y_train)}\nTest score : {classifier_knn.score(x_test, y_test)}")
```

```python
def svm_classifier(x_train, x_test, y_train, y_test):
    
    classifier_svm = SVC()
    classifier_svm.fit(x_train, y_train)

    y_pred = classifier_svm.predict(x_test)

    cm = confusion_matrix(y_test, y_pred)

    return print(f"Train score : {classifier_svm.score(x_train, y_train)}\nTest score : {classifier_svm.score(x_test, y_test)}")
```

```python
def print_score(X_train, X_test, y_train, y_test):
    print("Naivi Baise:\n")
    naive_classifier(X_train, X_test, y_train, y_test)

    print("-"*100)
    print()

    print("Random Forest:\n")
    random_classifier(X_train, X_test, y_train, y_test)

    print("-"*100)
    print()
    
    print("AdaBoost:\n")
    ada_classifier(X_train, X_test, y_train, y_test)

    print("-"*100)
    print()
    
    print("KNN:\n")
    knn_classifier(X_train, X_test, y_train, y_test)

    print("-"*100)
    print()
    
    print("SVM:\n")
    svm_classifier(X_train, X_test, y_train, y_test)

    print("-"*100)
    print()
```

```python
print_score(x_train, x_test, y_train, y_test)
```
